```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
  
data = pd.read_csv('data.csv') 
  
Y = data.iloc[0:,1].values  
R = data.iloc[0:,3].values  
D = data.iloc[0:,5].values  
X = data.iloc[0:,0]
  
plt.figure(figsize=(50,30)) 
  
ax = plt.axes() 
ax.grid(linewidth=0.9, color='#8f8f8f')  
  
ax.set_facecolor("black")  
ax.set_xlabel('\nTarih',size=25,color='#4bb4f2') 
ax.set_ylabel('Toplam Vaka Says\n', 
              size=25,color='#4bb4f2') 
  
plt.xticks(rotation='vertical',size='20',color='white') 
plt.yticks(size=20,color='white') 
plt.tick_params(size=20,color='white') 
  
for i,j in zip(X,Y): 
    ax.annotate(str(j),xy=(i,j+100),color='white',size='13') 
      
ax.annotate('Ykseli: 15th April', 
            xy=(15.2, 860), 
            xytext=(19.9,500), 
            color='white', 
            size='25', 
            arrowprops=dict(color='white', 
                            linewidth=0.025)) 


plt.title("COVID-19 Trkiye : Gnlk Vaka Deiimi\n", 
          size=50,color='#28a9ff') 
  
ax.plot(X,Y, 
        color='#1F77B4', 
        marker='o', 
        linewidth=4, 
        markersize=15, 
        markeredgecolor='#035E9B')
```
